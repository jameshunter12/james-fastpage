```python
import pandas as pd

df = pd.read_json("../football/footballdataset.json")

print(df)
```

```python
print(df[['Referee', 'AST', 'AS', 'AwayTeam', 'HomeTeam']].to_string(index=False))
```

```python
print(df[(df.HomeTeam == "Arsenal") | (df.AwayTeam == "Arsenal")])
```

```python
Arsenal_homesgames = df.loc[(df.HomeTeam == "Arsenal"), ['HomeTeam', 'FTHG', 'FTAG', 'AwayTeam', ]]
Arsenal_awaygames = df.loc[(df.AwayTeam == "Arsenal"), ['HomeTeam', 'FTHG', 'FTAG', 'AwayTeam',]]

print(Arsenal_homesgames.to_string(index=False))
print(Arsenal_awaygames.to_string(index=False))
```

```python
Arsenal_away_losses = Arsenal_awaygames.loc[Arsenal_awaygames.FTAG < Arsenal_awaygames.FTHG]
print(Arsenal_away_losses.to_string (index=False))
```
